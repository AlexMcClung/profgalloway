```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('https://github.com/AlexMcClung/profgalloway/raw/master/galloway-data.xlsx')
```

```
# need to transform value-to-cost ratio for plot - what type of transformation did Prof G use?
df['Value'] = np.log( df['Value-to-Cost Ratio'] )

# value vs vulnerability plot
plt.style.use('fivethirtyeight')
plt.figure(figsize=(16,10))

ax = sns.scatterplot(x='Value', y='Vulnerability Score', hue='Prof G Categorization', data=df)

ax.invert_yaxis()

plt.show()
```

```
plt.figure(figsize=(16,10))

df['Value-to-Cost Ratio'].plot(kind='box')

plt.show()
```

```
plt.figure(figsize=(16,10))

df['Vulnerability Score'].plot(kind='box')

plt.show()
```
